```python
import pandas as pd
```

```python
df = pd.read_csv('../data/temp/cobertura_movil_3Q2022.csv', delimiter=";")
```

```python
df
```

```python
# convertir pivot a long format
df_t = pd.melt(df, 
        id_vars=['nro', 'ubigeo', 'region', 'provincia', 'distrito', 'centro_poblado', 'lat', 'lon'], 
        value_vars=['2g', '3g', '4g', '5g']
        )
df_t = df_t[df_t['value']=='SI'].sort_values(by=['nro']).reset_index(drop=True)
df_t.drop(columns=['value'], inplace=True)
df_t.rename(columns={'variable': 'tecnologia'}, inplace=True)
```

```python
df_t = df_t[['lat', 'lon', 'tecnologia', 'centro_poblado', 'region']]
```

```python
#r'../data/entel_movil_parsed.json', 
df_t.to_json(r'../data/entel_movil_parsed.json', orient="records")
```
